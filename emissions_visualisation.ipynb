```python
monoterpene_min20 = pd.read_csv('output/F_monoterpene_min20.dat')[74:].mean()[0]
monoterpene_min15 = pd.read_csv('output/F_monoterpene_min15.dat')[74:].mean()[0]
monoterpene_min10 = pd.read_csv('output/F_monoterpene_min10.dat')[74:].mean()[0]
monoterpene_min5 = pd.read_csv('output/F_monoterpene_min5.dat')[74:].mean()[0]
monoterpene_0 = pd.read_csv('output/F_monoterpene_0.dat')[74:].mean()[0]
monoterpene_5 = pd.read_csv('output/F_monoterpene_5.dat')[74:].mean()[0]
monoterpene_10 = pd.read_csv('output/F_monoterpene_10.dat')[74:].mean()[0]
monoterpene_15 = pd.read_csv('output/F_monoterpene_15.dat')[74:].mean()[0]
monoterpene_20 = pd.read_csv('output/F_monoterpene_20.dat')[74:].mean()[0]
monoterpene_25 = pd.read_csv('output/F_monoterpene_25.dat')[74:].mean()[0]
monoterpene_30 = pd.read_csv('output/F_monoterpene_30.dat')[74:].mean()[0]
monoterpene_35 = pd.read_csv('output/F_monoterpene_35.dat')[74:].mean()[0]
monoterpene_40 = pd.read_csv('output/F_monoterpene_40.dat')[74:].mean()[0]

df_mean_mono_emissions_temp = [
    monoterpene_min20,
    monoterpene_min15,
    monoterpene_min10,
    monoterpene_min5,
    monoterpene_0,
    monoterpene_5,
    monoterpene_10,
    monoterpene_15,
    monoterpene_20,
    monoterpene_25,
    monoterpene_30,
    monoterpene_35,
    monoterpene_40
]
```

```python
df_temp = np.linspace(-20, 40, 13)
```

```python
plt.plot(df_temp, df_mean_iso_emissions_temp, marker='.', markersize=10)
plt.plot(df_temp, df_mean_mono_emissions_temp, marker='.', markersize=10)

plt.title("Emission rates as a function of temperature")
plt.xlabel("Temperature [deg C]")
plt.ylabel("Emissions")
plt.legend(["Isoprene", "Monoterpene"])
plt.grid()
plt.show()
```
