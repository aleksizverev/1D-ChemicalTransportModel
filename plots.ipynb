```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_time = pd.read_csv('output/time.dat')
df_particle_conc = pd.read_csv('output/particle_conc.dat', delimiter='\s+')
df_diam = pd.read_csv('output/diameter.dat')
```

```python
fig = plt.figure(1)
ax = plt.subplot(111)
X,Y = np.meshgrid(df_time[99:], df_diam[:45])

df_conc_cutted = df_particle_conc.iloc[100:, :44]*1e-6 # [# cm-3], number concentration
df_diam = df_diam[:44]
logDp = (np.log10(df_diam.values[1]*100) - np.log10(df_diam.values[0]*100))
df_dn_dlogdp = df_conc_cutted/logDp[0]

p = ax.pcolor(X, Y, df_dn_dlogdp.T, vmin=0, vmax=40000)
cb = fig.colorbar(p, ax=ax)
ax.set_yscale('log')
ax.set_ylabel('Diameter (m)')
ax.set_xlabel('Time (days)')
```

```python
plt.plot(df_diam[:50], df_dn_dlogdp.loc[106:115].T)
m, n = df_dn_dlogdp.loc[106:115].T.shape
plt.xscale('log')
plt.ylabel('dN/dLogDp (cm^-3)')
plt.grid(True, which="both")
plt.legend(np.arange(9, m, 1))
plt.xlabel("Diameter (m)")
plt.show()
```
